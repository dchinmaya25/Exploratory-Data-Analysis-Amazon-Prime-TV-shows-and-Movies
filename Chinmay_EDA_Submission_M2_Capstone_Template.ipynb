## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Copy the original dataframes to preserve raw data
titles = titles_df.copy()
credits = credits_df.copy()

# Handling missing values
# Replace missing 'age_certification' with 'Unknown'
titles['age_certification'] = titles['age_certification'].fillna('Unknown')

# Replace missing 'seasons' with 0 for movies (since only shows have seasons)
titles['seasons'] = titles['seasons'].fillna(0)

# Replace missing numeric values with median to avoid skew
num_cols = ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score', 'runtime']
for col in num_cols:
    titles[col] = titles[col].fillna(titles[col].median())

# Clean 'genres' and 'production_countries' columns
# Convert stringified lists to actual lists using ast.literal_eval (safe eval)
def clean_list_columns(x):
    if pd.isnull(x) or x == '[]' or x == '':
        return []
    try:
        return ast.literal_eval(x)
    except:
        return [x]

titles['genres'] = titles['genres'].apply(clean_list_columns)
titles['production_countries'] = titles['production_countries'].apply(clean_list_columns)

# Standardize text data
titles['type'] = titles['type'].str.strip().str.title()
titles['title'] = titles['title'].str.strip()
titles['age_certification'] = titles['age_certification'].str.strip()

# Remove duplicates if any
titles.drop_duplicates(subset='id', inplace=True)
credits.drop_duplicates(inplace=True)

# Verify the cleaned dataset
print("Data Wrangling Completed Successfully!\n")
print("Titles Dataset Shape:", titles.shape)
print("Credits Dataset Shape:", credits.shape)

# Check for remaining nulls
print("\nRemaining Missing Values in Titles Dataset:")
print(titles.isnull().sum()[titles.isnull().sum() > 0])

# Display cleaned data preview
display(titles.head())
````

### What all manipulations have you done and insights you found?



**Data Manipulations Performed**

Copied original dataframes (titles_df, credits_df) to preserve raw data and prevent accidental overwriting.

Handled missing values:

Replaced missing age_certification values with 'Unknown' (to retain all records).

Filled missing seasons with 0 (since only shows have seasons).

For numeric columns like runtime, imdb_score, imdb_votes, etc., filled missing values with the median to avoid skewing the data.

Cleaned list-like columns (genres, production_countries) using ast.literal_eval() to convert stringified lists into proper Python lists.

Standardized text columns (trimmed spaces and fixed capitalization) in columns like show_type, title, and age_certification for uniformity.

Removed duplicate records to ensure each title and credit entry is unique.

Verified data integrity by checking remaining missing values and dataset shapes after cleaning.

**Key Insights from Data Wrangling**

Some columns like age_certification and seasons had many missing values, mostly for older movies or special content.

Median imputation helped stabilize numeric distributions without distorting data.

Many titles have multiple genres and countries of production â€” indicating diverse international content.

After cleaning, the dataset is now consistent, structured, and ready for exploratory data analysis (EDA) and visualizations.
